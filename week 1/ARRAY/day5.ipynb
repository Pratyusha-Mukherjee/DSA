```python
#circular linked list doubly
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def traversefront(head):
    if head is None:
        print("List is empty")
        return

    current = head
    while True:
        print(current.data, end=" -> ")
        current = current.next
        if current == head:
            break
    print("null")




node1 = Node(10)
node2 = Node(67)
node3 = Node(79)
node4 = Node(2)

node1.next = node2
node2.prev = node1

node2.next = node3
node3.prev = node2

node3.next = node4
node4.prev = node3

node4.next = node1
node1.prev = node4

traversefront(node1)
```
