```python
#finding minimum value
array=[12,9,76,34,21]
min = array[0]
for i in array:
    if i < min:
        min = i

print("minimum value is:-",min)
```

```python
#bubble sort
arr=[12,9,76,34,21]
n = len(arr)
for i in range(n-1):
    for j in range(n-i-1):
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1]=arr[j+1],arr[j]

print("sorted array:",arr)
#complexity = O(n)^2
```

```python
#selection sort
arr = [12, 9, 76, 34, 21]
n = len(arr)
for i in range(n-1):
    min_index = i
    for j in range(i+1, n):
        if arr[j] < arr[min_index]:
            min_index = j
            """ this will create space complexity
              so we use swap algo instead of this
             min_val=arr.pop(min_index)  
             arr.insert(i,min_val)
             """
    arr[i], arr[min_index] = arr[min_index], arr[i]  # Move this line inside the outer loop

print("Sorted array:", arr)
#complexity = O(n)^2
```

```python
#insertion sort
arr = [12, 9, 76, 34, 21]
n = len(arr)
for i in range(1,n):
    insert_index = i
    current_val=arr[i]
    for j in range(i-1,-1,-1):
        if arr[j]>current_val:
            arr[j+1]=arr[j]
            insert_index = j
        else:
            break
    arr[insert_index]=current_val
 
print("Sorted array is:-",arr)            
#complexity = O(n)^2
```
